```python
import pandas as pd
import numpy as np
df = pd.read_csv('prices2022.csv')
a = [1.05 for i in range(12)]
df['hours']=pd.date_range(start = '20220101' , periods=len(df), freq = 'h')
df.set_index('hours', inplace = True)
dfmonths = df.resample('M').mean()
```

```python
df['months'] = df.index.month
df['h'] = df.index.hour
df = df[:'20221231']
dfmeans = df[df['h'] > 8]
dfmeans = dfmeans.groupby('months').mean()
dfmeans['monthly'] = list(df.resample('M').mean()['Price'])
dfmeans['ratio'] = dfmeans['Price']/dfmeans['monthly']
dfmeans
```

```python
dfmeans['new_rations'] = list(a)
```

```python
dfmeans['new_ration*monthly'] = dfmeans['new_rations']*dfmeans['monthly']
dfmeans['sub'] = dfmeans['new_ration*monthly']-dfmeans['Price']
```

```python
dfmeans.reset_index(0, inplace = True)
dfmeans = dfmeans[['months','sub']]
```

```python
df = df.merge(dfmeans, how = 'left',on = 'months')
```

```python
df['add'] = df['Price'] + df['sub']
df['substract'] = df['Price'] - (2*df['sub'])
df['newPrice'] = np.where((df.h >= 8 ),df['add'] ,df['substract'])
```

```python
df = df[['Price' , 'newPrice']]
df
```
